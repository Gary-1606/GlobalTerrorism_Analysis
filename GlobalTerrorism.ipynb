```python
#constructing data
def transformed_attack(attack):
    if attack == 'Unknown':
        return 0
    if attack == 'Armed Assault':
        return 1
    if attack == 'Bombing/Explosion':
        return 2
    if attack == 'Hostage Taking (Kidnapping)':
        return 3
    if attack == 'Hostage Taking (Barricade Incident)':
        return 4
    if attack == 'Hijacking':
        return 5
    if attack == 'Unarmed Assault':
        return 6
    if attack == 'Assassination':
        return 7
    if attack == 'Facility/Infrastructure Attack':
        return 8
```

# Treating outliers

```python
#outlier removal function based on quantile
def remove_outlier(df_in, col_name):
    q1 = df_in[col_name].quantile(0.25)
    q3 = df_in[col_name].quantile(0.75)
    iqr = q3-q1 #Interquartile range
    fence_low  = q1-1.5*iqr
    fence_high = q3+1.5*iqr
    df_out = df_in.loc[(df_in[col_name] > fence_low) & (df_in[col_name] < fence_high)]
    return df_out

outlier_terrorism = terrorismdf
outliers_terrorism = remove_outlier(outlier_terrorism, 'Year')
yeardf = outliers_terrorism
```
